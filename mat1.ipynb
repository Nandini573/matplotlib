# subplots

```python
fig, ax = plt.subplots(figsize=(15,16))
ax.scatter(batter['avg'],batter['strike_rate'],color='red',marker='+')
ax.set_title('something')
ax.set_xlabel('avg')
ax.set_ylabel('strike_rate')
fig.show()
```

```python
#2*1=2 graphs,2*2=4 graphs
fig, ax =plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[1].scatter(batter['avg'],batter['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
#3d surface plots
x =np.linspace(-10,10,100)
y =np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z =np.sin(xx)+np.cos(yy)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
#contour graph
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
